```python
lf_median = housing_data['LotFrontage'].median()

values = {'LotFrontage': lf_median, 'GarageYrBlt': 0, 'MasVnrArea': 0, 'Electrical': 'SBrkr'  }

housing_data.fillna(value=values, inplace = True)
housing_data.fillna(value = 'None', inplace = True)
```

```python
#change MSSUb class to object. It contains categories
housing_data['MSSubClass'] = housing_data['MSSubClass'].astype(np.str)
```

```python
#Create feature "HouseAge" with more correlation
housing_data['HouseAge'] = housing_data['YrSold'] - housing_data['YearRemodAdd']
#Create feature for "GarageAge"
housing_data['GarageAge'] = housing_data['YrSold'] - housing_data['GarageYrBlt'] #+ housing_data['MoSold']/12.0

#Garage cars correlated with Garage Area
housing_data.drop(['YrSold','YearRemodAdd', 'GarageCars'], axis = 1, inplace = True)

#drop Id, use pandas index, also drop year built as is similar to year_remodelled
housing_data.drop(['Id','YearBuilt', 'MoSold'], axis = 1, inplace = True)
```

```python
#Give weights to Garage Quality and Garage Condition, PoolQC
replacement = { "Ex": 5, "Gd":4, "TA":3, "Fa":2, "Po":1, "None": 0, "NA": 0 }

#replacement for paved-drive
replacement_pd = { "Y": 3, "P":2, "N": 1 } 

#replacement for basement finish
replacement_bsm = {"GLQ":6, "ALQ":5, "BLQ":4, "Rec":3, "LwQ":2, "Unf":1, "None": 0, "NA": 0 }

housing_data = housing_data.replace({'GarageQual': replacement, 'GarageCond':replacement, 'PoolQC':replacement,
                                     'FireplaceQu':replacement, 'PavedDrive':replacement_pd,
                                     'BsmtFinType1':replacement_bsm, 'BsmtFinType2':replacement_bsm
                                    })
housing_data['GarageOverallQual'] = housing_data['GarageQual'] *  housing_data['GarageCond'] * housing_data['GarageArea']
#housing_data['FireplaceOverall'] = housing_data['FireplaceQu'] * housing_data['Fireplaces']

housing_data['BsmtOverallQual1'] = housing_data['BsmtFinSF1'] * housing_data['BsmtFinType1']
housing_data['BsmtOverallQual2'] = housing_data['BsmtFinSF2'] * housing_data['BsmtFinType2']
```

```python
housing_data.drop(['BsmtFinSF2','BsmtFinType2', 'BsmtFinSF1', 'BsmtFinType1', 'GarageQual', 'GarageCond', 'GarageArea'], axis = 1, inplace = True)
```
